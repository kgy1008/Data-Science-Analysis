````python
#BusinessTravel(출장여부) 시각화
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
df['BusinessTravel'].value_counts().plot.pie(autopct='%1.1f%%')
plt.subplot(1,2,2)
sns.countplot(data=df, x='BusinessTravel')
df['BusinessTravel'].value_counts()
````

````python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
df['Department'].value_counts().plot.pie(autopct='%1.1f%%')
plt.subplot(1,2,2)
sns.countplot(data=df, x='Department')
df['Department'].value_counts()
````

````python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
df['EducationField'].value_counts().plot.pie(autopct='%1.1f%%')
plt.subplot(1,2,2)
sns.countplot(data=df, x='EducationField')
df['EducationField'].value_counts()
````

````python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%')
plt.subplot(1,2,2)
sns.countplot(data=df, x='Gender')
df['Gender'].value_counts()
````

````python
def a(s):
    plt.figure(figsize=(15,5))
    plt.subplot(1,2,1)
    df[s].value_counts().plot.pie(autopct='%1.1f%%')
    plt.subplot(1,2,2)
    sns.countplot(x = df[s])
    df[s].value_counts()

for i in df.columns:
    if df[i].dtypes=='O':
        print(i)
        a(i)
````

가설 설정 - 누가 퇴사를 할것인가?
